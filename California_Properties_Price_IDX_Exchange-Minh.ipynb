```python
missing_pct = mega_rc_filtered.isnull().mean().sort_values(ascending=False)
missing_pct
```

```python
# drop observations where missing_pct>0.99
drop_missing_cols = missing_pct[missing_pct > 0.99].index.tolist()
print(f">99% missing columns ({len(drop_missing_cols)})")
print(drop_missing_cols)
mega_rc_filtered = mega_rc_filtered.drop(columns=drop_missing_cols)

def bucket_missing_rate_v2(rate):
    # categorized based on mising ratio
    if rate >= 0.8:
        return 'High Missing (>80%)'
    elif rate >= 0.5:
        return 'Upper-Medium Missing (50%-80%)'
    elif rate >= 0.2:
        return 'Lower-Medium Missing (20%-50%)'
    else:
        return 'Low Missing (<20%)'

missing_df = missing_pct.drop(index = drop_missing_cols).reset_index()
missing_df.columns = ['Column', 'MissingRate']
missing_df['Bucket'] = missing_df['MissingRate'].apply(bucket_missing_rate_v2)

bucket_counts = missing_df['Bucket'].value_counts()

print("\nMissing Value Bucket Summary (4 Buckets):")
print(bucket_counts)

for bucket in ['High Missing (>80%)', 'Upper-Medium Missing (50%-80%)',
               'Lower-Medium Missing (20%-50%)', 'Low Missing (<20%)']:
    print(f"\n{bucket} columns:")
    display(missing_df[missing_df['Bucket'] == bucket][['Column', 'MissingRate']])
```

```python
#dropping columns that are more than 50% nulls
sparse = missing_pct[missing_pct > 0.5].index.tolist()

# also drop 0.5–0.9
#sparse = missing_prop[(missing_prop > 0.5) & (missing_prop <= 0.9)].index.tolist()

cols_to_drop = sparse #+ sparse
print(cols_to_drop)
#even though many seem logically relevant in pricing not enough data to include them in the model

mega_filtered_dropped = mega_rc_filtered.copy().drop(columns=cols_to_drop, errors='ignore')

missing_pct_fd = mega_filtered_dropped.copy().isnull().mean().sort_values(ascending=False) #new series of missingness percent
mega_fd = mega_filtered_dropped.copy()#copy of clean dataset to work with
mega_fd
```

```python
# Impute missing values in mega_fd
# You might want to adjust the imputation strategy based on the column's data type and distribution

# Separate columns by data type for different imputation strategies
numerical_cols = mega_fd.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = mega_fd.select_dtypes(include=['object', 'category', 'bool']).columns

# Impute numerical columns with the median (less sensitive to outliers than the mean)
for col in numerical_cols:
    if mega_fd[col].isnull().any():
        median_val = mega_fd[col].median()
        mega_fd[col].fillna(median_val, inplace=True)

# Impute categorical columns with the mode (most frequent value)
for col in categorical_cols:
    if mega_fd[col].isnull().any():
        mode_val = mega_fd[col].mode()[0] # .mode() can return multiple values if tied, take the first
        mega_fd[col].fillna(mode_val, inplace=True)

# Verify that there are no more missing values
print("Missing values after imputation:")
print(mega_fd.isnull().sum().sum())
```

```python
# Perform xgboost

import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error # Import MAPE
import pandas as pd

# Prepare dataset for XGBoost
# Use the same features and target as the last linear regression model
features_xgb = [
    'LivingArea', 'BedroomsTotal', 'BathroomsTotalInteger', 'YearBuilt',
    'LotSizeSquareFeet', 'PropertyType', 'PropertySubType', 'CountyOrParish',
    'City', 'Flooring', 'ViewYN', 'PoolPrivateYN', 'AttachedGarageYN',
    'FireplaceYN', 'Stories', 'Levels', 'NewConstructionYN', 'GarageSpaces',
    'HighSchoolDistrict', 'AssociationFee', 'PostalCode'
]
target_xgb = 'ClosePrice'

# Check if selected columns exist in the DataFrame
if not all(col in mega_fd.columns for col in features_xgb + [target_xgb]):
    print("Error: One or more specified columns not found in the DataFrame for XGBoost.")
else:
    X_xgb = mega_fd[features_xgb]
    y_xgb = mega_fd[target_xgb]

    # Identify categorical columns within the subset
    categorical_cols_xgb = X_xgb.select_dtypes(include=['object', 'category', 'bool']).columns

    # Apply one-hot encoding to categorical columns
    # handle_unknown='ignore' is useful if the test set might have categories not in the train set
    X_encoded_xgb = pd.get_dummies(X_xgb, columns=categorical_cols_xgb, dummy_na=False)

    # XGBoost can handle missing values internally, but since we imputed before,
    # we don't strictly need to drop NaNs here based on the imputed mega_fd.
    # If not imputing, you would need to decide how to handle NaNs for XGBoost.

    # Split the dataset into training and testing
    X_train_xgb, X_test_xgb, y_train_xgb, y_test_xgb = train_test_split(X_encoded_xgb, y_xgb, test_size=0.2, random_state=42)

    # Initialize and train the XGBoost Regressor model
    # You can tune these parameters later for better performance
    model_xgb = xgb.XGBRegressor(objective='reg:squarederror', # Regression task
                                 n_estimators=100,         # Number of boosting rounds
                                 learning_rate=0.1,        # Step size shrinkage
                                 max_depth=5,              # Maximum depth of a tree
                                 random_state=42,          # For reproducibility
                                 n_jobs=-1)               # Use all available cores

    print("\nTraining XGBoost model...")
    model_xgb.fit(X_train_xgb, y_train_xgb)
    print("XGBoost model training complete.")

    # Make predictions
    y_pred_xgb = model_xgb.predict(X_test_xgb)

    # Evaluate the model
    r2_xgb = r2_score(y_test_xgb, y_pred_xgb)
    mape_xgb = mean_absolute_percentage_error(y_test_xgb, y_pred_xgb) # Calculate MAPE

    print(f"\nXGBoost Regression Results for Features: {features_xgb} and Target: {target_xgb}")
    print(f"R-squared (R2): {r2_xgb:.2f}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape_xgb:.2f}") # Print MAPE

    # Optional: Print feature importances
    print("\nFeature Importances:")
    feature_importances = pd.Series(model_xgb.feature_importances_, index=X_train_xgb.columns).sort_values(ascending=False)
    display(feature_importances)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Prepare dataset for Lasso
# Use the same features and target as the last XGBoost model
features_lasso = [
    'LivingArea', 'BedroomsTotal', 'BathroomsTotalInteger', 'YearBuilt',
    'LotSizeSquareFeet', 'PropertyType', 'PropertySubType', 'CountyOrParish',
    'City', 'Flooring', 'ViewYN', 'PoolPrivateYN', 'AttachedGarageYN',
    'FireplaceYN', 'Stories', 'Levels', 'NewConstructionYN', 'GarageSpaces',
    'HighSchoolDistrict', 'AssociationFee', 'PostalCode'
]
target_lasso = 'ClosePrice'

# Check if selected columns exist in the DataFrame
if not all(col in mega_fd.columns for col in features_lasso + [target_lasso]):
    print("Error: One or more specified columns not found in the DataFrame for Lasso.")
else:
    X_lasso = mega_fd[features_lasso]
    y_lasso = mega_fd[target_lasso]

    # Identify categorical columns
    categorical_cols_lasso = X_lasso.select_dtypes(include=['object', 'category', 'bool']).columns

    # Apply one-hot encoding to categorical columns
    X_encoded_lasso = pd.get_dummies(X_lasso, columns=categorical_cols_lasso, dummy_na=False)

    # Scale the features
    scaler = StandardScaler()
    X_scaled_lasso = scaler.fit_transform(X_encoded_lasso)

    # Split the dataset into training and testing
    X_train_lasso, X_test_lasso, y_train_lasso, y_test_lasso = train_test_split(X_scaled_lasso, y_lasso, test_size=0.2, random_state=42)

    # Initialize and train the Lasso model
    # You can tune the alpha parameter (regularization strength) later
    lasso_model = Lasso(alpha=1.0, random_state=42) # alpha=1.0 is a common starting point

    print("\nTraining Lasso model...")
    lasso_model.fit(X_train_lasso, y_train_lasso)
    print("Lasso model training complete.")

    # Get the coefficients
    coefficients_lasso = pd.Series(lasso_model.coef_, index=X_encoded_lasso.columns)

    print("\nLasso Coefficients (indicating feature importance):")
    # Display non-zero coefficients
    display(coefficients_lasso[coefficients_lasso != 0].sort_values(ascending=False))
```
